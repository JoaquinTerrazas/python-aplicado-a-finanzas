# Construccion del Altman Z Score

#### Cinco ratios clave y pesos del modelo Z-score de Altman: • X1: Working Capital/Total Assets: 1.2 • X2: Retained Earnings/Total Assets: 1.4 • X3: EBIT/Total Assets: 3.3 • X4: Market Cap/Long-Term Liabilities: 0.6 • X5: Sales/Total Assets: 1

##### X1: Working capital/ total assets : 1.2

```python
# Primera opcion
print(yf.Ticker("ABT").balance_sheet.loc["Current Assets","2024-12-31"])
print(yf.Ticker("ABT").balance_sheet.loc["Current Liabilities","2024-12-31"])
print("Working Capital: USD",yf.Ticker("ABT").balance_sheet.loc["Current Assets","2024-12-31"]- yf.Ticker("ABT").balance_sheet.loc["Current Liabilities","2024-12-31"])
print("Working Capital: USD",yf.Ticker("ABT").balance_sheet.loc["Working Capital", "2024-12-31"])
```

```python
ratios = {"Ticker":[],"Nombre":[],"Mkt_cap":[],"WK":[],"Total_Assets":[],"Retained Earnings":[],"EBIT":[],"LT_Debt":[],"Revenue":[]}
```

```python
for i in tickers_health_ss:    # Recorre cada ticker
    df=yf.Ticker(i)             # Crea el objeto Ticker de yfinance para ese símbolo
    ratios["Ticker"].append(i)    # Agrega el ticker a la columna 'Ticker'
    ratios["Nombre"].append(df.info["longName"])    # Agrega el nombre largo desde el dict df.info
    ratios["Mkt_cap"].append(df.info["marketCap"])    # Agrega la capitalización bursátil (marketCap) desde df.info
    ratios["WK"].append(df.balance_sheet.loc["Working Capital",df.balance_sheet.columns[0]])   # Agrega Working Capital desde el balance,  # Usa la 1ª columna (suele ser el período más reciente)
    ratios["Total_Assets"].append(df.balance_sheet.loc["Total Assets",df.balance_sheet.columns[0]])  # Total de activos del último período
    ratios["Retained Earnings"].append(df.balance_sheet.loc["Retained Earnings",df.balance_sheet.columns[0]])    # Utilidades retenidas del último período
    ratios["EBIT"].append(df.income_stmt.loc["EBIT",df.income_stmt.columns[0]])  # EBIT desde el estado de resultados
    ratios["LT_Debt"].append(df.balance_sheet.loc["Long Term Debt And Capital Lease Obligation",df.balance_sheet.columns[0]])   # 1ª columna (período más reciente)
    ratios["Revenue"].append(df.income_stmt.loc["Total Revenue",df.income_stmt.columns[0]])  

ratios   # Muestra el diccionario ya llenado
df_ratios=pd.DataFrame(ratios).sort_values(by="Mkt_cap", ascending=False)    # Crea DataFrame y ordena por Mkt_cap desc.  
df_ratios   # Muestra la tabla final
 
```

```python
df_ratios["X1"]=df_ratios["WK"]/df_ratios["Total_Assets"]  # Capital de trabajo / Activos totales
df_ratios["X2"]=df_ratios["Retained Earnings"]/df_ratios["Total_Assets"]     # Utilidades retenidas / Activos totales
df_ratios["X3"]=df_ratios["EBIT"]/df_ratios["Total_Assets"]    # EBIT / Activos totales
df_ratios["X4"]=df_ratios["Mkt_cap"]/df_ratios["LT_Debt"]       # Valor de mercado / Deuda LP
df_ratios["X5"]=df_ratios["Revenue"]/df_ratios["Total_Assets"]   # Ventas (Revenue) / Activos totales
```

```python
df_ratios
```

### Fórmula de Altman Z-Score
Z= 1.2X1 + 1.4X2 + 3.3X3 + 0.6X4 + 1X5.

Esta herramienta clasifica a las empresas en las siguientes categorías según
su puntuación Z:
- Zona en dificultades (Z < 1.81): indica alto riesgo de quiebra. Se
recomienda precaución.
- Zona gris (1.81<Z<3): muestra una posición financiera precaria. Se
recomienda vigilancia.
- Zona Segura (Z > 3): indica estabilidad y bienestar financiero.

```python
df_ratios["Z"]=1.2*df_ratios["X1"]+1.4*df_ratios["X2"]+3.3*df_ratios["X3"]+0.6*df_ratios["X4"]+1*df_ratios["X5"]
df_ratios
```

```python
# funcion para clasficar la empresa segun el caluclo de su ratio de Altman Z-Score
def altman_z_score_classification(z):
    if z>3.0:
        return "Saludable"
    elif z>1.81:
        return "En vigilancia"
    else:
        return "En riesgo de quiebra"
```

```python
altman_z_score_classification(3.05)
```

```python
df_ratios["Z_Classification"]=df_ratios["Z"].apply(altman_z_score_classification)
df_ratios
```
