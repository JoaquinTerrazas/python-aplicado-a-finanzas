```python
# Seleccion de fechas especificas
ticker=["AAPL","MSFT"]
start_date = "2018-12-30"
end_date = "2025-08-12"
```

```python
datos = yf.download(ticker, start=start_date, end= end_date, auto_adjust=False)["Adj Close"]
datos  #lo baja como un dataframe ya que yfinance ya tiene incorporado pandas
```

```python
#importamos matplotlib
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.plot(datos.index, datos.values,label=f"{ticker} Adj Close", linewidth=2) #en el eje x se ponen fechas(datos.index) y eje y se ponen los precios (datos.value)
plt.grid(True) # para las lineas del graficos (los cuadrantes) y linewidt para que las lineas de la serie sean mas gruesas
plt.show
```

```python
datos["AAPL"].plot(figsize=(10,5), title="AAPL")
plt.show()
```

```python
datos["MSFT"].plot(figsize=(10,5), title="MSFT")
plt.show()
```

```python
datos["AAPL"].plot(figsize= (10, 5))
datos["MSFT"].plot(figsize= (10, 5))
plt.title("Evolucion de los precios de AAPL y MSFT")
plt.text(datos.index[-1],datos["AAPL"].iloc[-1],round(datos["AAPL"].iloc[-1],1))  #plt.text (x, y, text).  # datos.index[-1] la utlima fecha de la serie.   # datos["AAPL"] la columna con precios AAPL
plt.text(datos.index[-1],datos["MSFT"].iloc[-1],round(datos["MSFT"].iloc[-1],1))  # .iloc[-1] el ultimo valor de esa columna.  # round(datos["AAPL"].iloc[-1], 1)  redondea el ultimo valor a un decimal
plt.xlabel("Fecha") 
plt.ylabel("Precio")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
for t in ticker:   #recorre cada elemento t dentro de ticker
    plt.plot(datos.index, datos[t], label=t, linewidth=2)  # plt.plot (x, y, ...) dibuja una linea.  # datos.index en eje x fechas.  # datos[t] en eje y serie de precios del ticker t
    # anotar el último valor de cada serie
    plt.text(datos.index[-1], datos[t].iloc[-1], f"{t}: {datos[t].iloc[-1]:.1f}") # plt.text(x, y, texto) escribe texto dentro del gráfico.  # datos.index[-1] → última fecha.  
# datos[t].iloc[-1] → último valor de la serie del ticker t.  # f"{t}: {datos[t].iloc[-1]:.1f}" → f-string que arma el texto: {t} inserta el ticker. {datos[t].iloc[-1]:.1f} formatea el número con 1 decimal.                                                                         
plt.title("Evolución de precios ajustados: AAPL vs MSFT")
plt.xlabel("Fecha")
plt.ylabel("Precio (USD)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
