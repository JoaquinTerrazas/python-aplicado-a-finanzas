```python
import numpy as np
import matplotlib.pyplot as plt
```

**Simulación de un Geometric Brownian Motion de la forma**

$dS_t = \mu S_t dt + \sigma S_t dW_t$

Donde:
- $S_t$: Precio del activo en el tiempo $t$.
- $\mu$: Tasa de crecimiento esperada (drift).
- $\sigma$: Volatilidad.
- $dW_t$: Incremento del movimiento browniano.

La solución analítica para simular trayectorias es:

$S_t = S_0 e^{(\mu - \frac{\sigma^2}{2})t + \sigma W_t}$

#**Vamos a simular una cuantas trayectorias aleatorios y graficar el promedio de acuerdo a la fórmula**

```python
# Parámetros de la simulación
S0 = 3.5        # Precio inicial del activo
mu = 0.08        # Tasa de retorno esperada (anual)
sigma = 0.02     # Volatilidad (anual)
T = 2.0         # Horizonte temporal (en años)
N = 252*2         # Número de pasos (por ejemplo, 252 días hábiles en un año)
dt = T / N      # Tamaño del paso temporal
n_simulations = 100 # Número de trayectorias a simular
```

```python
# Generar trayectorias del GBM
t = np.linspace(0, T, N+1)  # Vector de tiempos
S = np.zeros((N+1, n_simulations))  # Matriz para almacenar las trayectorias
S[0, :] = S0  # Establecer el precio inicial
```

```python
# Simular las trayectorias
for i in range(N):
    # Generar incrementos del movimiento browniano
    Z = np.random.normal(0, 1, n_simulations)  # Variables aleatorias normales
    # Calcular el precio en el siguiente paso
    S[i+1, :] = S[i, :] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)
```

```python
# Graficar las trayectorias
plt.figure(figsize=(10, 6))
for j in range(n_simulations):
    plt.plot(t, S[:, j], label=f'Trayectoria {j+1}' if j < 5 else None)  # Mostrar etiquetas solo para las primeras 5
# Añadir el precio esperado (tendencia determinística)
S_expected = S0 * np.exp(mu * t)
plt.plot(t, S_expected, 'k--', label='Tendencia esperada (E[S_t])', linewidth=2)

# Configurar la gráfica
plt.xlabel('Tiempo (años)', fontsize=12)
plt.ylabel('Precio del activo (S_t)', fontsize=12)
plt.title('Simulación de Geometric Brownian Motion', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()

# Imprimir algunos parámetros para referencia
print(f"Parámetros de la simulación:")
print(f"Precio inicial (S0): {S0}")
print(f"Tasa de retorno esperada (mu): {mu}")
print(f"Volatilidad (sigma): {sigma}")
print(f"Horizonte temporal (T): {T} años")
print(f"Número de pasos (N): {N}")
print(f"Tamaño del paso (dt): {dt:.6f}")
```

**Hay veces en las que las trayectorias no están alrededor del promedio, para ellas usamos más trayectorias y vemos que en promedio están alrededor del promedio**

```python
# Parámetros de la simulación
S0 = 100        # Precio inicial del activo
mu = 0.18        # Tasa de retorno esperada (anual)
sigma = 0.30     # Volatilidad (anual)
T = 2.0         # Horizonte temporal (en años)
N = 252*2         # Número de pasos (por ejemplo, 252 días hábiles en un año)
dt = T / N      # Tamaño del paso temporal
n_simulations = 100  # Número de trayectorias a simular

# Generar trayectorias del GBM
t = np.linspace(0, T, N+1)  # Vector de tiempos
S = np.zeros((N+1, n_simulations))  # Matriz para almacenar las trayectorias
S[0, :] = S0  # Establecer el precio inicial

# Simular las trayectorias
for i in range(N):
    # Generar incrementos del movimiento browniano
    Z = np.random.normal(0, 1, n_simulations)  # Variables aleatorias normales
    # Calcular el precio en el siguiente paso
    S[i+1, :] = S[i, :] * np.exp((mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z)

# Graficar las trayectorias
plt.figure(figsize=(10, 6))
for j in range(n_simulations):
    plt.plot(t, S[:, j], label=f'Trayectoria {j+1}' if j < 5 else None)  # Mostrar etiquetas solo para las primeras 5

# Añadir el precio esperado (tendencia determinística)
S_expected = S0 * np.exp(mu * t)
plt.plot(t, S_expected, 'k--', label='Tendencia esperada (E[S_t])', linewidth=2)

# Configurar la gráfica
plt.xlabel('Tiempo (años)', fontsize=12)
plt.ylabel('Precio del activo (S_t)', fontsize=12)
plt.title('Simulación de Geometric Brownian Motion', fontsize=14)
plt.grid(True)
plt.legend()
plt.show()

# Imprimir algunos parámetros para referencia
print(f"Parámetros de la simulación:")
print(f"Precio inicial (S0): {S0}")
print(f"Tasa de retorno esperada (mu): {mu}")
print(f"Volatilidad (sigma): {sigma}")
print(f"Horizonte temporal (T): {T} años")
print(f"Número de pasos (N): {N}")
print(f"Tamaño del paso (dt): {dt:.6f}")
```
