```python
def get_row_val(df, row_names):
    """
    Devuelve el primer valor disponible (columna más reciente) para
    cualquiera de los nombres en row_names. Si no existe, NaN.
    """
    if df is None or df.empty:
        return np.nan
    if isinstance(row_names, str):
        row_names = [row_names]
    for rn in row_names:
        if rn in df.index:
            # toma la columna más reciente (suele ser la primera)
            return df.loc[rn].iloc[0]
    return np.nan

rows = []
for t in tickers_energy:
    tk = yf.Ticker(t)
    info = getattr(tk, "info", {}) or {}
    bs = getattr(tk, "balance_sheet", pd.DataFrame())
    is_ = getattr(tk, "income_stmt", pd.DataFrame())

    row = {
        "Ticker": t,
        "Nombre": info.get("longName", np.nan),
        "Mkt_cap": info.get("marketCap", np.nan),

        # Balance general
        "WK": get_row_val(bs, ["Working Capital"]),  # a veces no existe
        "Total_Assets": get_row_val(bs, ["Total Assets"]),
        "Retained Earnings": get_row_val(bs, ["Retained Earnings"]),
        "LT_Debt": get_row_val(
            bs,
            ["Long Term Debt", "Long Term Debt And Capital Lease Obligation"]
        ),

        # Estado de resultados
        "EBIT": get_row_val(is_, ["EBIT", "Ebit"]),
        "Revenue": get_row_val(is_, ["Total Revenue", "TotalRevenue"]),
    }
    rows.append(row)

df_ratios = pd.DataFrame(rows).sort_values(by="Mkt_cap", ascending=False)
df_ratios
```

```python
df_ratios["X1"]=df_ratios["WK"]/df_ratios["Total_Assets"]
df_ratios["X2"]=df_ratios["Retained Earnings"]/df_ratios["Total_Assets"]
df_ratios["X3"]=df_ratios["EBIT"]/df_ratios["Total_Assets"]
df_ratios["X4"]=df_ratios["Mkt_cap"]/df_ratios["LT_Debt"]
df_ratios["X5"]=df_ratios["Revenue"]/df_ratios["Total_Assets"]
```

```python
df_ratios
```

```python
df_ratios["Z"]=1.2*df_ratios["X1"]+1.4*df_ratios["X2"]+3.3*df_ratios["X3"]+0.6*df_ratios["X4"]+1*df_ratios["X5"]
df_ratios
```

```python
def altman_z_score_classification(z):
  if z>3.0:
    return "Saludable"
  elif z>1.81:
    return "En vigilancia"
  else:
    return "En riesgo de quiebra"
```

```python
df_ratios["Z_Classification"]=df_ratios["Z"].apply(altman_z_score_classification)
df_ratios
```
